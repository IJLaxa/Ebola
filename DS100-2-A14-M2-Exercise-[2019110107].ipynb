```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows',None)
data = pd.read_csv('https://raw.githubusercontent.com/IJLaxa/Ebola/main/ebola_2014_2016_clean.csv')
Country_data = data.loc[data['Country'].str.contains('Nigeria|Liberia',regex=True)]
Country_data
```

```python
Country_data = Country_data.loc[Country_data['Date'].str.contains('2014-08|2014-12')]
Country_data
```

```python
Country_data['Date'] = pd.to_datetime(Country_data['Date'])
desired_months = ['August', 'September', 'October', 'November', 'December']
Country_data = Country_data.loc[(Country_data['Date'].dt.strftime('%B').isin(desired_months)) & (Country_data['Date'].dt.year == 2014)]
Country_data['Date'] = Country_data['Date'].dt.strftime('%Y-%m-%d')
Country_data
```

```python
Country_data['CFR'] = np.round(data['Cumulative no. of confirmed, probable and suspected deaths']/data['Cumulative no. of confirmed, probable and suspected cases']*100,3)
Country_data
```

```python
df1 = Country_data.loc[Country_data['Country'].str.contains('Nigeria')]
df2 = Country_data.loc[Country_data['Country'].str.contains('Liberia')]
df1 = (df1.rename(columns={'Country': 'Country1','CFR':'CFR1'}))
df2 = (df2.rename(columns={'Country': 'Country2','CFR':'CFR2'}))
df1 = df1.drop(columns=['Cumulative no. of confirmed, probable and suspected cases','Cumulative no. of confirmed, probable and suspected cases'])
df2 = df2.drop(columns=['Cumulative no. of confirmed, probable and suspected cases','Cumulative no. of confirmed, probable and suspected deaths'])
abs_data = pd.merge(df1,df2)
abs_data[['Date','Country1','CFR1','Country2','CFR2']]
```

```python
abs_data['Absolute Difference'] = np.round(abs(abs_data['CFR1']-abs_data['CFR2']),3)
abs_data[['Date','Country1','CFR1','Country2','CFR2','Absolute Difference']]
```

Daily New Cases

```python
%matplotlib notebook
data=pd.read_csv('ebola_2014_2016_clean.csv')
df_NIR=data.loc[data['Country'].str.contains('Nigeria',regex=True)]
df_LIBR=data.loc[data['Country'].str.contains('Liberia',regex=True)]
df=pd.concat([df_NIR,df_LIBR])

dailynewcases=df_NIR[['Date','Cumulative no. of confirmed, probable and suspected cases']]
graph=dailynewcases.plot.bar('Date','Cumulative no. of confirmed, probable and suspected cases',rot=30,color='blue',width=1)
plt.title('Daily New Cases in the Nigeria')
plt.xlabel('Dates')
plt.ylabel('Number of Cases')
plt.tight_layout()
ax = plt.gca()
ax.locator_params(nbins=12, axis='x')

dailynewcases=df_LIBR[['Date','Cumulative no. of confirmed, probable and suspected cases']]
graph=dailynewcases.plot.bar('Date','Cumulative no. of confirmed, probable and suspected cases',rot=30,color='green',width=1)
plt.title('Daily New Cases in the Liberia')
plt.xlabel('Dates')
plt.ylabel('Number of Cases')
plt.tight_layout()
ax = plt.gca()
ax.locator_params(nbins=12, axis='x')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import pytz
ebola_df = pd.read_csv('ebola_2014_2016_clean.csv', parse_dates=['Date'])
nigeria_df = ebola_df[ebola_df['Country'] == 'Nigeria']
lagos_tz = pytz.timezone('Africa/Lagos')
nigeria_df['Date'] = pd.to_datetime(nigeria_df['Date']).dt.tz_localize(pytz.utc).dt.tz_convert(lagos_tz)
nigeria_df['CFR'] = nigeria_df['Cumulative no. of confirmed, probable and suspected deaths'] / nigeria_df['Cumulative no. of confirmed, probable and suspected cases']
fig, ax = plt.subplots()
ax.plot(nigeria_df['Date'], nigeria_df['CFR'])
ax.set_title('Ebola Outbreak in Nigeria (2014-2016)')
ax.set_xlabel('Date')
ax.set_ylabel('Case Fatality Ratio')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


ebola_df = pd.read_csv('ebola_2014_2016_clean.csv')
nigeria_df = ebola_df[ebola_df['Country'] == 'Nigeria']


nigeria_df['CFR'] = nigeria_df['Cumulative no. of confirmed, probable and suspected deaths'] / nigeria_df['Cumulative no. of confirmed, probable and suspected cases']


first_date = pd.to_datetime(nigeria_df['Date'].min())
nigeria_df['Days'] = (pd.to_datetime(nigeria_df['Date']) - first_date).dt.days


fig = plt.figure()
ax = plt.axes(projection='3d')


x = nigeria_df['Days']
y = nigeria_df['Cumulative no. of confirmed, probable and suspected cases']
z = nigeria_df['CFR']


ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')


ax.set_title('Case Fatality Rate (CFR) of Ebola in Nigeria')
ax.set_xlabel('Days since First Date')
ax.set_ylabel('Cumulative Number of Cases')
ax.set_zlabel('CFR')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# Load the cleaned Ebola dataset for Liber
ebola_df = pd.read_csv('ebola_2014_2016_clean.csv')
Liberia_df = ebola_df[ebola_df['Country'] == 'Liberia']

# Compute the case fatality rate (CFR)
Liberia_df['CFR'] = Liberia_df['Cumulative no. of confirmed, probable and suspected deaths'] / Liberia_df['Cumulative no. of confirmed, probable and suspected cases']

# Convert the Date column to days since the first date
first_date = pd.to_datetime(Liberia_df['Date'].min())
Liberia_df['Days'] = (pd.to_datetime(Liberia_df['Date']) - first_date).dt.days

# Create a 3D figure
fig = plt.figure()
ax = plt.axes(projection='3d')

# Define the x, y, and z values
x = Liberia_df['Days']
y = Liberia_df['Cumulative no. of confirmed, probable and suspected cases']
z = Liberia_df['CFR']

# Plot the 3D surface
ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')

# Set the plot title and axis labels
ax.set_title('Case Fatality Rate (CFR) of Ebola in Liberia')
ax.set_xlabel('Days since First Date')
ax.set_ylabel('Cumulative Number of Cases')
ax.set_zlabel('CFR')

# Show the plot
plt.show()
```

Animation

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation


ebola_df = pd.read_csv('ebola_2014_2016_clean.csv', parse_dates=['Date'])


Nigeria_df = ebola_df[ebola_df['Country'] == 'Nigeria']


nigeria_df['CFR'] = Nigeria_df['Cumulative no. of confirmed, probable and suspected deaths'] /Nigeria_df['Cumulative no. of confirmed, probable and suspected cases']


fig, ax = plt.subplots()


def update(num):
  
    ax.clear()

    
    ax.plot(nigeria_df['Date'].iloc[:num], nigeria_df['CFR'].iloc[:num], color='blue')

   
    ax.set_xlabel('Date')
    ax.set_ylabel('Case Fatality Ratio')
    ax.set_title('Ebola Outbreak in Nigeria (2014-2016)')


ani = animation.FuncAnimation(fig, update, frames=nigeria_df.shape[0], interval=100)


plt.show()
```

Discussion: The data used in this Matplotlib shows the case fatality rate (CFR) of Ebola in Liberia over time, based on the cleaned Ebola dataset for 2014-2016. The CFR is computed as the ratio of the cumulative number of deaths to the cumulative number of confirmed, probable, and suspected cases.The plot shows the relationship between the number of days since the first date of the Ebola outbreak in Liberia, the cumulative number of cases, and the CFR. The x-axis shows the number of days since the first date of the Ebola outbreak, the y-axis shows the cumulative number of cases, and the z-axis shows the CFR. For example the 3D surface plot is colored using the "viridis" colormap, which maps low values to blue and high values to yellow.

The plot reveals some interesting insights. For example, we can see that the CFR was initially high when the outbreak started, but it decreased over time as more cases were reported and better treatment methods were developed. We can also see that there were periods of time when the CFR increased temporarily, which may have been due to factors such as changes in the virus or delays in treatment.Overall, this Matplotlib 3D plot is a useful tool for visualizing the relationship between multiple variables in the Ebola dataset for Liberia. It allows us to see trends and patterns that may not be apparent from a 2D plot, and it provides insights into the dynamics of the Ebola outbreak in Liberia over time.
